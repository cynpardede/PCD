```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Membaca gambar grayscale
image = cv2.imread("dede.png", cv2.IMREAD_GRAYSCALE)

# Menampilkan citra asli
plt.figure(figsize=(8,6))
plt.title("Citra Asli")
plt.imshow(image, cmap='gray')
plt.axis('off')
plt.show()

# ==========================
# Sobel Edge Detection
# ==========================

sobel_x = cv2.Sobel(
    image,
    cv2.CV_64F,
    1,
    0,
    ksize=3
)

sobel_y = cv2.Sobel(
    image,
    cv2.CV_64F,
    0,
    1,
    ksize=3
)

sobel_combined = cv2.magnitude(
    sobel_x,
    sobel_y
)

sobel_combined = np.uint8(
    sobel_combined /
    sobel_combined.max() * 255
)

plt.figure(figsize=(8,6))
plt.title("Deteksi Tepi Sobel")
plt.imshow(sobel_combined, cmap='gray')
plt.axis('off')
plt.show()

# ==========================
# Canny Edge Detection
# ==========================

edges_canny = cv2.Canny(
    image,
    100,
    200
)

plt.figure(figsize=(8,6))
plt.title("Deteksi Tepi Canny")
plt.imshow(edges_canny, cmap='gray')
plt.axis('off')
plt.show()

# ==========================
# Laplacian Edge Detection
# ==========================

laplacian = cv2.Laplacian(
    image,
    cv2.CV_64F
)

laplacian = np.uint8(
    np.absolute(laplacian)
)

plt.figure(figsize=(8,6))
plt.title("Deteksi Tepi Laplacian")
plt.imshow(laplacian, cmap='gray')
plt.axis('off')
plt.show()

print("Praktikum 8 berhasil dijalankan")
```
