```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Membaca gambar
image = cv2.imread("dede.png")

# Konversi BGR ke RGB
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Menampilkan citra asli
plt.figure(figsize=(8,6))
plt.title("Citra Asli")
plt.imshow(image_rgb)
plt.axis('off')
plt.show()

# ==========================
# Segmentasi Warna (HSV)
# ==========================

hsv_image = cv2.cvtColor(
    image,
    cv2.COLOR_BGR2HSV
)

lower_red = np.array([0,100,100])
upper_red = np.array([10,255,255])

mask = cv2.inRange(
    hsv_image,
    lower_red,
    upper_red
)

segmented_image = cv2.bitwise_and(
    image_rgb,
    image_rgb,
    mask=mask
)

plt.figure(figsize=(8,6))
plt.title("Segmentasi Warna Merah")
plt.imshow(segmented_image)
plt.axis('off')
plt.show()

# ==========================
# Thresholding
# ==========================

gray_image = cv2.cvtColor(
    image,
    cv2.COLOR_BGR2GRAY
)

_, binary_image = cv2.threshold(
    gray_image,
    127,
    255,
    cv2.THRESH_BINARY
)

plt.figure(figsize=(8,6))
plt.title("Segmentasi Thresholding")
plt.imshow(binary_image, cmap='gray')
plt.axis('off')
plt.show()

# ==========================
# Edge Detection
# ==========================

edges = cv2.Canny(
    gray_image,
    100,
    200
)

plt.figure(figsize=(8,6))
plt.title("Segmentasi Canny")
plt.imshow(edges, cmap='gray')
plt.axis('off')
plt.show()

print("Praktikum 7 berhasil dijalankan")
```
