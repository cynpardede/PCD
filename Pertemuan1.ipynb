```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Membaca gambar yang diupload
image = cv2.imread("dede.png")

# Konversi BGR ke RGB
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Menampilkan citra asli
plt.figure(figsize=(8,6))
plt.imshow(image_rgb)
plt.title("Citra Asli")
plt.axis('off')
plt.show()

# Konversi ke grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Menampilkan citra grayscale
plt.figure(figsize=(8,6))
plt.imshow(gray_image, cmap='gray')
plt.title("Citra Grayscale")
plt.axis('off')
plt.show()

# Resize menjadi 300x300/;
resized_image = cv2.resize(image_rgb, (300,300))

# Menampilkan hasil resize
plt.figure(figsize=(6,6))
plt.imshow(resized_image)
plt.title("Citra Resize 300x300")
plt.axis('off')
plt.show()

# Menyimpan hasil
cv2.imwrite(
    "processed_image.jpg",
    cv2.cvtColor(resized_image, cv2.COLOR_RGB2BGR)
)

print("Proses selesai! File disimpan sebagai processed_image.jpg")
```
