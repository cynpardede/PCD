```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Membaca gambar grayscale
image = cv2.imread("dede.png", cv2.IMREAD_GRAYSCALE)

# Menampilkan citra asli
plt.figure(figsize=(8,6))
plt.title("Citra Asli")
plt.imshow(image, cmap='gray')
plt.axis('off')
plt.show()

# ==========================
# Thresholding
# ==========================

_, binary_image = cv2.threshold(
    image,
    127,
    255,
    cv2.THRESH_BINARY
)

plt.figure(figsize=(8,6))
plt.title("Citra Biner")
plt.imshow(binary_image, cmap='gray')
plt.axis('off')
plt.show()

# ==========================
# Kernel Morfologi
# ==========================

kernel = np.ones((5,5), np.uint8)

# ==========================
# Erosion
# ==========================

eroded_image = cv2.erode(
    binary_image,
    kernel,
    iterations=1
)

plt.figure(figsize=(8,6))
plt.title("Hasil Erosion")
plt.imshow(eroded_image, cmap='gray')
plt.axis('off')
plt.show()

# ==========================
# Dilation
# ==========================

dilated_image = cv2.dilate(
    binary_image,
    kernel,
    iterations=1
)

plt.figure(figsize=(8,6))
plt.title("Hasil Dilation")
plt.imshow(dilated_image, cmap='gray')
plt.axis('off')
plt.show()

# ==========================
# Opening
# ==========================

opened_image = cv2.morphologyEx(
    binary_image,
    cv2.MORPH_OPEN,
    kernel
)

plt.figure(figsize=(8,6))
plt.title("Hasil Opening")
plt.imshow(opened_image, cmap='gray')
plt.axis('off')
plt.show()

# ==========================
# Closing
# ==========================

closed_image = cv2.morphologyEx(
    binary_image,
    cv2.MORPH_CLOSE,
    kernel
)

plt.figure(figsize=(8,6))
plt.title("Hasil Closing")
plt.imshow(closed_image, cmap='gray')
plt.axis('off')
plt.show()

print("Praktikum 9 berhasil dijalankan")
```
