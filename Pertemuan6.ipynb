```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Membaca gambar grayscale
image = cv2.imread("dede.png", cv2.IMREAD_GRAYSCALE)

# Menampilkan citra asli
plt.figure(figsize=(8,6))
plt.title("Citra Asli")
plt.imshow(image, cmap='gray')
plt.axis('off')
plt.show()

# ======================
# Low-Pass Filter
# Gaussian Blur
# ======================

blurred_image = cv2.GaussianBlur(
    image,
    (5,5),
    0
)

plt.figure(figsize=(8,6))
plt.title("Low-Pass Filter (Gaussian Blur)")
plt.imshow(blurred_image, cmap='gray')
plt.axis('off')
plt.show()

# ======================
# Sobel Edge Detection
# ======================

sobel_x = cv2.Sobel(
    image,
    cv2.CV_64F,
    1,
    0,
    ksize=3
)

sobel_y = cv2.Sobel(
    image,
    cv2.CV_64F,
    0,
    1,
    ksize=3
)

sobel_combined = cv2.magnitude(
    sobel_x,
    sobel_y
)

sobel_combined = np.uint8(
    sobel_combined /
    sobel_combined.max() * 255
)

plt.figure(figsize=(8,6))
plt.title("Sobel Edge Detection")
plt.imshow(sobel_combined, cmap='gray')
plt.axis('off')
plt.show()

# ======================
# Canny Edge Detection
# ======================

edges = cv2.Canny(
    image,
    100,
    200
)

plt.figure(figsize=(8,6))
plt.title("Canny Edge Detection")
plt.imshow(edges, cmap='gray')
plt.axis('off')
plt.show()

print("Praktikum 6 berhasil dijalankan")
```
