```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Membaca gambar yang diupload
image = cv2.imread("dede.png")

# Konversi BGR ke RGB
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Menampilkan citra asli
plt.figure(figsize=(8,6))
plt.title("Citra Asli")
plt.imshow(image_rgb)
plt.axis('off')
plt.show()

# =====================
# Translasi
# =====================

def translate_image(image, x_shift, y_shift):
    translation_matrix = np.float32([
        [1,0,x_shift],
        [0,1,y_shift]
    ])

    return cv2.warpAffine(
        image,
        translation_matrix,
        (image.shape[1], image.shape[0])
    )

translated_image = translate_image(
    image_rgb,
    50,
    30
)

plt.figure(figsize=(8,6))
plt.title("Citra Hasil Translasi")
plt.imshow(translated_image)
plt.axis('off')
plt.show()

# =====================
# Rotasi
# =====================

(h,w) = image_rgb.shape[:2]
center = (w//2, h//2)

rotation_matrix = cv2.getRotationMatrix2D(
    center,
    45,
    1.0
)

rotated_image = cv2.warpAffine(
    image_rgb,
    rotation_matrix,
    (w,h)
)

plt.figure(figsize=(8,6))
plt.title("Citra Hasil Rotasi")
plt.imshow(rotated_image)
plt.axis('off')
plt.show()

# =====================
# Flipping
# =====================

flip_horizontal = cv2.flip(image_rgb,1)
flip_vertical = cv2.flip(image_rgb,0)

plt.figure(figsize=(12,6))

plt.subplot(1,2,1)
plt.imshow(flip_horizontal)
plt.title("Flip Horizontal")
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(flip_vertical)
plt.title("Flip Vertikal")
plt.axis('off')

plt.show()

# =====================
# Scaling
# =====================

new_size = (
    int(w*0.7),
    int(h*0.7)
)

scaled_image = cv2.resize(
    image_rgb,
    new_size
)

plt.figure(figsize=(8,6))
plt.title("Citra Hasil Scaling")
plt.imshow(scaled_image)
plt.axis('off')
plt.show()

print("Praktikum 4 berhasil dijalankan")
```
